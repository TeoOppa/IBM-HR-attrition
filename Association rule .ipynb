```python
del (df['Unnamed: 0'])
```

```python
df['AgeBin'] = pd.cut(df['Age'].astype(int), 10, right=False)

df.drop(['Age'], axis=1, inplace=True)

df.head(100)
```

```python
df.head(100)
```

```python
df['Attrition'] = df['Attrition'].map(
    {0: 'No', 1: 'Yes'}).astype(str)
df['BusinessTravel'] = df['BusinessTravel'].map(
    {0: 'Non-Travel', 1: 'Travel_Rarely', 2: 'Travel_Frequently'}).astype(str)
df['Department'] = df['Department'].map(
    {0: 'Sales', 1: 'Research & Development', 2: 'Human Resources'}).astype(str)
df['EducationField'] = df['EducationField'].map(
    {0: 'Human Resources', 1: 'Life Sciences', 2: 'Marketing', 3: 'Medical', 4:'Other', 5: 'Technical Degree'}).astype(str)
df['Gender'] = df['Gender'].map(
    {0: 'Female', 1: 'Male'}).astype(str)

df['AgeBin'] = df['AgeBin'].astype(str) + '_Age'


df.head()
```

```python
del (df['Attrition_numerical'])
```
